# *Visualization of Snowflakes*

```python
G = nx.Graph()
```

```python
G.add_node("fact_sales")
```

```python
G.add_node("dim_product")
G.add_node("dim_customer")
G.add_node("dim_time")

G.add_node("product_category")
G.add_node("product_brand")

G.add_node("customer_region")
G.add_node("customer_city")

G.add_node("year")
G.add_node("month")

G.add_edge("fact_sales", "dim_product")
G.add_edge("fact_sales", "dim_customer")
G.add_edge("fact_sales", "dim_time")

G.add_edge("dim_product", "product_category")
G.add_edge("dim_product", "product_brand")

G.add_edge("dim_customer", "customer_region")
G.add_edge("customer_region", "customer_city")

G.add_edge("dim_time", "year")
G.add_edge("year", "month")
```

```python
plt.figure(figsize=(8,8))
pos = nx.spring_layout(G, seed=42)
```

```python
nx.draw(G, pos, with_labels=True, node_color='skyblue',
        node_size=2000, font_size=9, font_weight='bold')

plt.title("Snowflake Schema Representation")
plt.show()
```
